**VISUALIZATION OF CONTINENTS V/S AVERAGE POPULATION IN 2023 OF EACH CONTINENT**

```python
avg_pop = df.groupby('continent')['2023 population'].mean().reset_index()
avg_sort = avg_pop.sort_values(by='2023 population', ascending=False)
```

```python
colors = ['#350387', '#3f1387', '#52249e', '#6d38c4', '#8b4df2']

plt.figure(figsize=(12, 6))
ax = plt.axes()
ax.set_facecolor("#a89fb6")
plt.xticks(rotation=45) 
plt.bar(avg_sort['continent'], avg_sort['2023 population'], color=colors)
plt.xlabel('Continents',fontsize=15)
plt.ylabel(f'Average Population in 2023', fontsize=13)
plt.title(f'Average Population by Continent in 2023')
plt.show()
```
